```python
filename = 'candidateTotalMoneySwap.txt'
df = pd.read_csv(filename, dtype=object, delimiter='|')
df
```

```python
matr_df = df.ix[:, '1':'24']
matr_df
```

```python
matr = matr_df.as_matrix()
matr
```

```python
xs, ys = np.triu_indices_from(matr)
zip(xs,ys)
```

```python
for i, j in zip(xs, ys):
    if i != j:
        print '"%s", "%s", "%s"' %(df['Unnamed: 0'].ix[i], df['Unnamed: 0'].ix[j], matr[i][j])
```

```python
df_manual = pd.read_csv('candidates_analysis.csv', delimiter=';', index_col='names').reindex(df['Unnamed: 0'].as_matrix())
df_manual = df_manual.reset_index()
df_manual
```

```python
df_manual
```

```python
matr_df_manual = df_manual[['taxes', 'banking policy', 'government regulations', 'international trade', 'budgets',
                    'labor and employment', 'Iran nuclear deal', 'healthcare', 'immigration',
                    'abortion', 'gay rights']]
matr_df_manual
```

```python
np_matr_df_manual = matr_df_manual.as_matrix()
np_matr_df_manual
```

```python
for (x, y) in itertools.combinations(xrange(0,24), 2):
    dist = 0.0
    count = 0
    for j in xrange(0, 11):
        if  not np.isnan(np_matr_df_manual[x][j]) and not np.isnan(np_matr_df_manual[y][j]):
            dist += abs(np_matr_df_manual[x][j] - np_matr_df_manual[y][j])
            count += 1
    if count > 0:
        dist = dist / count
    else:
        dist = 100
    print '"%s", "%s", "%s"' %(df_manual['names'].ix[x], df_manual['names'].ix[y], dist)
```
