```python
trimmed_zip = name_zip_empl_digit_zip.copy()
trimmed_zip['contbr_zip'] = trimmed_zip['contbr_zip'].map(lambda z : z[:5] if len(z) > 5 else z)
trimmed_zip.head()
```

```python
counted_by_zip = trimmed_zip.groupby(['contbr_zip'])
```

```python
def f(grp):
    #print grp.shape
    to_replace = []
    for x, y in itertools.combinations(grp.values, 2):
        if x != y:
            if x in y:
                to_replace.append((x, y))
            elif y in x:
                to_replace.append((y,x))
            elif editdistance.eval(x, y) <= 2:
                to_replace.append((x, y) if len(x) > len(y) else (y, x))
    
    for x, y in to_replace:
        grp.replace(x, y, inplace=True)
    return grp
```

```python
name_da = trimmed_zip.copy()
name_da['contbr_nm_da'] = counted_by_zip['contbr_nm'].transform(f)
```
